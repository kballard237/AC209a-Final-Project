```python
import numpy as np
import pandas as pd
import os
import datetime
from matplotlib import pyplot as plt
```

```python
def read_txt(filename):
    save_ls = []
    with open(filename, 'r') as file:
        for line in file:
            line = line.split(',')
            for val in line:
                save_ls.append(val)
    return save_ls

# read in sensor data
hum = read_txt('./collection1/CS109a_humidity.txt')
light = read_txt('./collection1/CS109a_light.txt')
motion = read_txt('./collection1/CS109a_motion.txt')
pressure = read_txt('./collection1/CS109a_pressure.txt')
temp = read_txt('./collection1/CS109a_temperature.txt')

# generate times (every 30 minutes) from initial start date
oct23 = datetime.datetime(2019, 10, 23, hour=8, minute=16)
dates = [oct23]
for i in range(73): 
    dates.append(dates[i] + datetime.timedelta(minutes=30))
    
# create dataframe of variables
df = pd.DataFrame({'date': dates,'temp': temp, 'hum': hum, 'press': pressure, 'light': light,
                   'motion': motion})
# drop last empty row
df.drop(df.tail(1).index,inplace=True)
```
