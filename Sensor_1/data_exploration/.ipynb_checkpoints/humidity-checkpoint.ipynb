```python
import numpy as np
import pandas as pd
import os
import datetime
from matplotlib import pyplot as plt
```

```python
def read_txt(filename):
    save_ls = []
    with open(filename, 'r') as file:
        for line in file:
            line = line.split(',')
            for val in line:
                save_ls.append(val)
    return save_ls

# read in sensor data
data_1 = read_txt('../data_1/CS109a_humidity.txt') # 30 min, 48 hours
data_2 = read_txt('../data_2/CS109a_humidity.txt') # 30 min, 48 hours
data_3 = read_txt('../data_3/CS109a_humidity.txt') # 30 min, 24 hours
data_4 = read_txt('../data_4/CS109a_humidity.txt') # 10 min, 48 hours
data_5 = read_txt('../data_5/CS109a_humidity.txt') # 10 min, 72 hours
data_6 = read_txt('../data_6/CS109a_humidity.txt')
data_7 = read_txt('../data_7/CS109a_humidity.txt')
data_8 = read_txt('../data_8/CS109a_humidity.txt')
data_9 = read_txt('../data_9/CS109a_humidity.txt')
data_10 = read_txt('../data_10/CS109a_humidity.txt')
data_11 = read_txt('../data_11/CS109a_humidity.txt')
```

```python
# check data length
data_list = [data_1, data_2, data_3, data_4, data_5, data_6, data_7, data_8, data_9, data_10, data_11] 

for i in range(len(data_list)):
    print(len(data_list[i]))
```

```python
# generate start times (every 30 minutes) for first 3, then every 10) from initial start dates
dates_1 = datetime.datetime(2019, 10, 29, hour=22, minute=36)
dates_2 = datetime.datetime(2019, 11, 1, hour=9, minute=14)
dates_3 = datetime.datetime(2019, 11, 6, hour=20, minute=15)
dates_4 = datetime.datetime(2019, 11, 8, hour=9, minute=5)
dates_5 = datetime.datetime(2019, 11, 10, hour=13, minute=51)
dates_6 = datetime.datetime(2019, 11, 13, hour=18, minute=45)
dates_7 = datetime.datetime(2019, 11, 16, hour=20, minute=56)
dates_8 = datetime.datetime(2019, 11, 19, hour=21, minute=57) 
dates_9 = datetime.datetime(2019, 11, 26, hour=19, minute=42)
dates_10 = datetime.datetime(2019, 11, 29, hour=23, minute=1)
dates_11 = datetime.datetime(2019, 12, 2, hour=23, minute=22)



def compute_time_int(start_dt, hum, freq):
    dates = [start_dt]
    for i in range(len(hum)-1): 
        dates.append(dates[i] + datetime.timedelta(minutes=freq))
    df = pd.DataFrame({'date': dates, 'hum': hum})
    df.drop(df.tail(1).index,inplace=True)
    return df
    
df0 = compute_time_int(dates_1, data_1, 30)
df1 = compute_time_int(dates_2, data_2, 30)
df2 = compute_time_int(dates_3, data_3, 30)
df3 = compute_time_int(dates_4, data_4, 10)
df4 = compute_time_int(dates_5, data_5, 10)
df5 = compute_time_int(dates_6, data_6, 10)
df6 = compute_time_int(dates_7, data_7, 10)
df7 = compute_time_int(dates_8, data_8, 10)
df8 = compute_time_int(dates_9, data_9, 10)
df9 = compute_time_int(dates_10, data_10, 10)
df10 = compute_time_int(dates_11, data_11, 10)
# drop last empty row
```

```python
def convert_to_float(dfs):
    for df in dfs:
        df.hum = [float(i) for i in df.hum]

def fill_in_dates(start_dt, end_dt):
    in_between = []
    cur_dt = start_dt
    in_between.append(cur_dt)
    while cur_dt <= end_dt:
        cur_dt += datetime.timedelta(minutes=5)
        in_between.append(cur_dt)
    return in_between

def create_large_df(dfs, ls=None, fill_in=False):
    if fill_in:
        b0 = fill_in_dates(max(df0.date), min(df1.date))
        b1 = fill_in_dates(max(df1.date), min(df2.date))
        b2 = fill_in_dates(max(df2.date), min(df3.date))
        b3 = fill_in_dates(max(df3.date), min(df4.date))

        df11 = pd.DataFrame({'date': b1, 'hum': float(l2[-2])*np.ones(len(b1))})
        df21 = pd.DataFrame({'date': b2, 'hum': float(l3[-2])*np.ones(len(b2))})
        df31 = pd.DataFrame({'date': b3, 'hum': float(l4[-2])*np.ones(len(b3))})

        df = pd.concat(dfs, axis=0)
    else:
        df = pd.concat(dfs, axis=0)
    return df

def assign_numerical_date_vals(df):
    df2 = df.copy()
    
    month = []
    day = []
    hour = []
    minute = []
    weekday = []
    
    for i,dt in enumerate(df2['date']):
        day.append(dt.day)
        hour.append(dt.hour)
        minute.append(dt.minute)
        month.append(dt.month)
        weekday.append(dt.weekday())

    df2['day'] = day
    df2['hour'] = hour
    df2['minute'] = minute
    df2['month'] = month
    df2['weekday'] = weekday
    return df2

def get_day_labels(df):
    days = df['day']
    unique_days =list(np.unique(days))
    indexes = np.unique(days, return_index=True)[1]
    dates_in_order = [days.values[index] for index in sorted(indexes)]
    return dates_in_order

def calendar_plot(df, dates_in_order, feature):
    fig, ax = plt.subplots(6,6, figsize=(18,15), sharey=True)
    ax = ax.flatten()

    for i,d in enumerate(dates_in_order):
        temp = df[df['day'] == d]
        ax[i].scatter(temp['hour'], temp[feature])
        ax[i].set_title("Day {}".format(d))
    plt.suptitle(feature)
    plt.xlabel("Hour")
    plt.ylabel(feature)
    plt.tight_layout()
    plt.show()
```

```python
dfs = [df0, df1, df2, df3, df4, df5, df6, df8, df9, df10]   

# convert values to floats
convert_to_float(dfs)

# create large df from all collection samples
dfs_to_concat = dfs


hum = create_large_df(dfs_to_concat)  

# plot based on calendar day
hum2 = assign_numerical_date_vals(hum)
dates_in_order = get_day_labels(hum2)
calendar_plot(hum2, dates_in_order, 'hum')
```

```python
def ave_hum(x):
    x['avg_hrly_hum'] = np.mean(x['hum'])
    return x

hum2 = hum2.reset_index()
del hum2['index']

hum3 = hum2.groupby(['month','day', 'hour']).apply(ave_hum)
hum4 = hum3.drop(columns=['date', 'minute', 'hum'])
hum5 = hum4.drop_duplicates(subset=["month","day", 'hour'])
```
