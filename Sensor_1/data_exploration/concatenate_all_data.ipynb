```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
hum = pd.read_csv("humidity_df.csv").drop(columns=['Unnamed: 0'])
temp = pd.read_csv("temp_df.csv").drop(columns=['Unnamed: 0'])
press = pd.read_csv("pressure_df.csv").drop(columns=['Unnamed: 0'])
motion = pd.read_csv("motion_df.csv").drop(columns=['Unnamed: 0'])
light = pd.read_csv("light_df.csv").drop(columns=['Unnamed: 0'])
```

```python
data = [hum, temp, press, motion, light]
for i in data:
    print(i.columns, i.shape)
```

```python
df1 = hum.merge(temp, left_on=['day', 'hour', 'month','weekday'], right_on=['day', 'hour', 'month','weekday'])
df2 = df1.merge(press, left_on=['day', 'hour', 'month','weekday'], right_on=['day', 'hour', 'month','weekday'])
df3 = df2.merge(motion, left_on=['day', 'hour', 'month','weekday'], right_on=['day', 'hour', 'month','weekday'])
df4 = df3.merge(light, left_on=['day', 'hour', 'month','weekday'], right_on=['day', 'hour', 'month','weekday'])
```

```python
df4.to_csv("all_sensor_data.csv")
```
