### Functions

```python
def read_txt(filename):
    """
    A function that opens text files and appends to list.
    
    Inputs:
        filename: files to load
    Outputs:
        save_ls: list of files 
    """
    
    save_ls = []
    with open(filename, 'r') as file:
        for line in file:
            line = line.split(',')
            for val in line:
                save_ls.append(val)
                
    return save_ls

###########################################################

def load_data(path):
    """
    A function that loads text files using read_text function and 
    changes list of entries to floats.
    
    Inputs:
        path: files to open
    Outputs:
        data: list of data as floats
    """
    
    data = read_txt(path)
    data = [float(data[i]) for i in range(len(data)-1)]
    
    return data

###########################################################

def create_timestamp(data, collection_freq, start_year, start_month, start_day, start_hour, start_min):
    
    """
    A function that creates timestamps for each data entry
    based on start time and collection frequency.
    
    Inputs:
        data: list of data values
        collection_freq: user-set frequency
        start_year: year sensor was started (i.e. 2019
        start_month: month number was started (i.e. 10 = Oct)
        start_day: number day sensor was started
        start_hour: hour of day sensor was started
        start_min: minute of day sensor was started
    Outputs:
        dates: timestamp object for each data entry
    """
    
    start_date = datetime.datetime(start_year, start_month, start_day, hour=start_hour, minute=start_min)
    dates = [start_date]
    
    for i in range(len(data)-1): 
        dates.append(dates[i] + datetime.timedelta(minutes=collection_freq))
    
    return dates

###########################################################

def build_df(dates_list, data_list, var):
    
    """
    A function that combines data from all collections 
    into single dataframe.
    
    Inputs:
        dates_list: list of all collection dates
        data_list: list of all collection data of var
        var: string of variable name

    Outputs:
        df: dataframe with data values and corresponding timestamps
    """
    
    df = pd.DataFrame()
    
    for i in range(len(data_list)):
        df_temp = pd.DataFrame({'date': dates_list[i], var : data_list[i]})
        df = df.append(df_temp, ignore_index = True)
        
    return df

###########################################################

def build_df_mult_vars(dates_list, data_list1, data_list2, data_list3, var1, var2, var3):
    
    """
    A function that combines data from all collections 
    into single dataframe.
    
    Inputs:
        dates_list: list of all collection dates
        data_list1: list of all collection data of var 1
        data_list2: list of all collection data of var 2
        data_list3: list of all collection data of var 3
        var1: string of variable name 1
        var2: string of variable name 2
        var3: string of variable name 3

    Outputs:
        df: dataframe with data values and corresponding timestamps
    """
    
    df = pd.DataFrame()
    
    for i in range(len(data_list1)):
        df_temp = pd.DataFrame({'date': dates_list[i], var1 : data_list1[i], 
                                var2 : data_list2[i], var3 : data_list3[i]})
        df = df.append(df_temp, ignore_index = True)
        
    return df

###########################################################

def assign_numerical_date_vals(df):
    df2 = df.copy()
    
    month = []
    day = []
    hour = []
    minute = []
    second = []
    weekday = []
    
    for i,dt in enumerate(df2['date']):
        month.append(dt.month)
        day.append(dt.day)
        hour.append(dt.hour)
        minute.append(dt.minute)
        second.append(dt.second)
        weekday.append(dt.weekday())

    df2['month'] = month
    df2['day'] = day
    df2['hour'] = hour
    df2['minute'] = minute
    df2['second'] = second
    df2['weekday'] = weekday
    return df2
```

# Visualizing Issues

**Fig 1:** Temperature and humidity recording failure after blank collection on Nov 23

```python
### Load data ###

# Temperature
temp_1 = load_data('data_1\CS109a_temperature.txt')
temp_2 = load_data('data_2\CS109a_temperature.txt')
temp_3 = load_data('data_3\CS109a_temperature.txt')
temp_4 = load_data('data_4\CS109a_temperature.txt')
temp_5 = load_data('data_5\CS109a_temperature.txt')
temp_6 = load_data('data_6\CS109a_temperature.txt')
temp_7 = load_data('data_7\CS109a_temperature.txt')
temp_8 = load_data('data_8\CS109a_temperature.txt')
temp_9 = load_data('data_9\CS109a_temperature.txt')
temp_10 = load_data('data_10\CS109a_temperature.txt')
temp_11 = load_data('data_11\CS109a_temperature.txt')

# Pressure
press_1 = load_data('data_1\CS109a_pressure.txt')
press_2 = load_data('data_2\CS109a_pressure.txt')
press_3 = load_data('data_3\CS109a_pressure.txt')
press_4 = load_data('data_4\CS109a_pressure.txt')
press_5 = load_data('data_5\CS109a_pressure.txt')
press_6 = load_data('data_6\CS109a_pressure.txt')
press_7 = load_data('data_7\CS109a_pressure.txt')
press_8 = load_data('data_8\CS109a_pressure.txt') 
press_9 = load_data('data_9\CS109a_pressure.txt')
press_10 = load_data('data_10\CS109a_pressure.txt')
press_11 = load_data('data_11\CS109a_pressure.txt')

# Humidity
hum_1 = load_data('data_1\CS109a_humidity.txt')
hum_2 = load_data('data_2\CS109a_humidity.txt')
hum_3 = load_data('data_3\CS109a_humidity.txt')
hum_4 = load_data('data_4\CS109a_humidity.txt')
hum_5 = load_data('data_5\CS109a_humidity.txt')
hum_6 = load_data('data_6\CS109a_humidity.txt')
hum_7 = load_data('data_7\CS109a_humidity.txt')
hum_8 = load_data('data_8\CS109a_humidity.txt') 
hum_9 = load_data('data_9\CS109a_humidity.txt')
hum_10 = load_data('data_10\CS109a_humidity.txt')
hum_11 = load_data('data_11\CS109a_humidity.txt')
```

```python
### Create timestamps for each data point ###

collection_freq0 = 30 # 1 samp/ 30 mins
collection_freq = 10  # 1 samp/ 10 mins

dates_1 = create_timestamp(temp_1, collection_freq0, 2019, 10, 29, 22, 36)
dates_2 = create_timestamp(temp_2, collection_freq0, 2019, 11, 1, 9, 14)
dates_3 = create_timestamp(temp_3, collection_freq0, 2019, 11, 6, 20, 15)
dates_4 = create_timestamp(temp_4, collection_freq, 2019, 11, 8, 9, 5)
dates_5 = create_timestamp(temp_5, collection_freq, 2019, 11, 10, 13, 51)
dates_6 = create_timestamp(temp_6, collection_freq, 2019, 11, 13, 18, 45)
dates_7 = create_timestamp(temp_7, collection_freq, 2019, 11, 16, 20, 56)
dates_8 = create_timestamp(temp_8, collection_freq, 2019, 11, 19, 21, 57) 
dates_9 = create_timestamp(temp_9, collection_freq, 2019, 11, 26, 19, 42)
dates_10 = create_timestamp(temp_10, collection_freq, 2019, 11, 29, 23, 1)
dates_11 = create_timestamp(temp_11, collection_freq, 2019, 12, 2, 23, 22)


### Create data frame ###

dates_list = [dates_1, dates_2, dates_3, dates_4, dates_5, 
              dates_6, dates_7, dates_8, dates_9, dates_10, dates_11] 
temp_list = [temp_1, temp_2, temp_3, temp_4, temp_5, temp_6, 
             temp_7, temp_8, temp_9, temp_10, temp_11] 
press_list = [press_1, press_2, press_3, press_4, press_5, 
               press_6, press_7, press_8, press_9, press_10, press_11] 
hum_list = [hum_1, hum_2, hum_3, hum_4, hum_5, hum_6, hum_7, 
            hum_8, hum_9, hum_10, hum_11] 

df = build_df_mult_vars(dates_list, temp_list, press_list, hum_list, 'temperature', 'pressure', 'humidity')


### Normalize data for easier visualization ###

cols_to_norm = ['temperature','pressure','humidity'] 
df[cols_to_norm] = MinMaxScaler(feature_range=(0,1)).fit_transform(df[cols_to_norm])


### Break down date into month, day, and other time components ###

df2 = assign_numerical_date_vals(df)
df2.head()
```

```python
## Add null rows for missing days for visualization ###

new_row1 = {'month': 11, 'day': 4}
new_row2 = {'month': 11, 'day': 5}
new_row3 = {'month': 11, 'day': 23}
new_row4 = {'month': 11, 'day': 24}
new_row5 = {'month': 11, 'day': 25}

df3 = df2.append(new_row1, ignore_index=True)
df3 = df3.append(new_row2, ignore_index=True)
df3 = df3.append(new_row3, ignore_index=True)
df3 = df3.append(new_row4, ignore_index=True)
df3 = df3.append(new_row5, ignore_index=True)

### Group data by month and day for visualization ###
grouped = df3.groupby(['month', 'day']).groups

dates = []
indices = []
for key in grouped.keys():
    dates.append(key)
    indices.append(grouped[key])
    
### Plot temperature, pressure and humidity data like a calendar ###
fig, ax = plt.subplots(6,7, figsize=(20,18), sharey=True)
ax = ax.flatten()

for i,d in enumerate(dates):
    rows_df = df3[(df3['month'] == d[0]) & (df3['day'] == d[1])]
    ax[i+2].scatter(rows_df['hour'], rows_df['temperature'], label = 'temp')
    ax[i+2].scatter(rows_df['hour'], rows_df['pressure'], label = 'press')
    ax[i+2].scatter(rows_df['hour'], rows_df['humidity'], label = 'humid')
    ax[i+2].legend()
    ax[i+2].set_xlim([0,24])
    ax[i+2].set_xlabel('Hour of Day')
    ax[i+2].set_ylabel('Data Value')
    if d[0] == 10:
        ax[i+2].set_title('Oct {}'.format(int(d[1])), fontsize = 15)
    if d[0] == 11:
        ax[i+2].set_title('Nov {}'.format(int(d[1])), fontsize = 15)
    if d[0] == 12:
        ax[i+2].set_title('Dec {}'.format(int(d[1])), fontsize = 15)
plt.tight_layout()
plt.suptitle('Temperature, Pressure, and Humidity Data per Collection Day', fontsize = 20)
fig.subplots_adjust(top=0.92)
fig.savefig('Sensor1_Issues_TempPressHum.png')
plt.show()
```
